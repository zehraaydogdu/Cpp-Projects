# C++ Dair Bazı Notlar  

C++'da bir değişkeni veya fonksiyonu kullanmadan önce derleyiciye (compiler) bunun ne olduğunu söylemek zorundasın. Derleyici, kodunu makine diline çevirirken her bir varlığın bellekte ne kadar yer kaplayacağını ve o veriyle hangi işlemlerin yapılabileceğini bilmek ister.
Eğer bir değişkeni int (tam sayı) olarak tanımladıysan, o değişken bellekte bir tam sayı kutusu olarak ayrılır. Programın ilerleyen satırlarında aynı değişken ismine "merhaba" gibi bir metin (string) atayamazsın.

## C++ Veri Türü Kategorileri  

### 1. Temel (Yerleşik) Türler - Fundamental Types
Bunlar dilin çekirdeğinde bulunan, doğrudan donanım tarafından anlaşılan türlerdir.  
- Tam Sayılar (int, long ....)  
- Karakterler (char ...)  
- Mantıksal (bool ...)  
- Ondalıklı (float, double, long ...)  
- Boş tür (void)  

### 2. Bileşik Türler - Compound Types  
Temel türlerin bir araya getirilmesi veya onlara referans verilmesiyle oluşurlar.
- Diziler (Arrays): Aynı tipten verilerin yan yana sıralanmasıdır (örn: int sayilar[5]).
- İşaretçiler (Pointers): Bir verinin değerini değil, bellekteki adresini tutarlar (örn: int* ptr).
- Referanslar (References): Mevcut bir değişkenin "takma adı" gibi davranırlar (örn: int& ref = x).
- Fonksiyonlar: Belirli bir işi yapan kod bloklarıdır ve bir dönüş tipine sahiptirler.

### 3. Kullanıcı Tanımlı Türler - User-Defined Types  
Programcının kendi ihtiyacına göre yarattığı karmaşık yapılardır.
- Struct: Farklı veri tiplerini bir arada tutan basit yapılar.
- Class (Sınıf): Nesne yönelimli programlamanın temeli. Hem veriyi hem de o veriyle yapılacak işlemleri (metotları) içerir.
- Enum: Belirli sabit değerleri isimlendirmek için kullanılır (örn: Günler, Renkler).
- Union: Aynı bellek alanını farklı veri tipleriyle kullanmanıza izin veren özel yapılar.

### Auto Kavramı  
Auto, bir değişkene ilk değerini atarken (initialization) değişkenin türünü sizin yazmanıza gerek kalmadan derleyicinin otomatik olarak tahmin etmesini sağlar.
. Örneğin auto a = 1 + 2; yazdığınızda derleyici sonucun tam sayı olacağını bilir ve a'yı int yapar.  

### Uzay Gemisi Operatörü (Spaceship Operator <=>)  
C++20 standardıyla dile eklenen yepyeni bir üçlü karşılaştırma (three-way comparison) operatörüdür
. Şeklinden dolayı bu lakabı almıştır. İki nesneyi birbiriyle kıyaslamayı inanılmaz derecede basitleştirir.
. C programlama dilindeki ünlü strcmp fonksiyonu gibi çalışır: İki nesneyi kıyaslar ve kıyas sonucunun 0, pozitif veya negatif bir tam sayı olmasına dayanarak bir sonuç (object) döndürür
. Karşılaştırma operatörlerinin programcılar tarafından kodlanmasını aşırı derecede kolaylaştırır.  

### Sizeof() ve size_t farkı  
- Sizeof: Bir operatördür. Amacı boyut hesaplamaktır. Değeri derleyici tarafından hesaplanan bir sayıdır. sizeof(int).  
- size_t: İşaretsiz bir veri tipidir. Amacı hesaplanan boyutu veya indeksi saklamaktır. Değeri değişkenin içine koyacağın herhangi bir pozitif tam sayıdır. size_t sayac = 0; .

İşaretli sayıların kullanılması isteniyor. İşaretsizleri çok özel durumlarda kullanmalı deniyor.  

### IEEE 754 Standardı ve Bellek Temsili (Representation)
Modern bilgisayarlar ve C++, ondalıklı sayıları matematikteki gibi dümdüz tutamaz; bunun yerine 1985'te ortaya çıkan ve 2019'da güncellenen IEEE 754 standardını kullanır.
Bu standarda göre ondalıklı sayılar bellekte bilimsel gösterime benzer şekilde 3 parçaya ayrılarak tutulur:
- İşaret Biti (Sign): Sayının pozitif mi negatif mi olduğunu belirten 1 bit  
- Üs (Exponent): Sayının büyüklüğünü belirler
  float için 8 bit, double için 11 bit ayrılmıştır.
- Mantisa (Mantissa / Significant): Sayının virgülden sonraki hassasiyetini belirten ana değerdir
  float için 23 bit, double için 52 bit ayrılmıştır.  

### Özel Değerler: Normal, Denormal, Sonsuz ve NaN
Bellekte tutulan sayıların değerlerine göre bazı "özel durumlar" oluşur:
- Normal ve Denormal Sayılar: Normal sayılar, bilgisayarın standart olarak temsil edebildiği değerlerdir. Ancak sıfıra çok ama çok yaklaştığımızda (örneğin 10 üzeri −40 gibi), standart gösterim yetersiz kalır. Burada Denormal (veya subnormal) sayılar devreye girer. Bunlar, sıfır etrafındaki hassasiyet boşluğunu doldurmak için kullanılan, üssü tamamen sıfırlardan oluşan çok ama çok küçük sayılardır.  
- Sonsuzluk (Infinity - inf): Bir işlemin sonucu, o veri tipinin alabileceği maksimum değeri aşarsa (taşma/overflow olursa) C++'ta tam sayılardaki gibi "Tanımsız Davranış" oluşmaz. Bunun yerine sonuç doygunluğa ulaşır ve inf (sonsuzluk) değerini alır
Örneğin 5.0 / 0.0 yaparsanız program çökmez, ekrana inf yazdırır
Sonsuzluğu başka bir normal sayıyla çarpar veya toplarsanız sonuç yine sonsuz çıkar
- Sayı Değil (Not a Number - NaN): Tanımsız veya matematikte karşılığı olmayan işlemleri temsil eder
Örneğin 0.0 / 0.0, negatif sayının karekökü veya sonsuzdan sonsuzu çıkarma (inf - inf) gibi işlemler NaN üretir. İşin en garip tarafı şudur: NaN hiçbir şeye eşit değildir, kendisine bile. C++'ta (NaN == NaN) ifadesi false döner.  

### Makine Epsilonu ve ULP (Hassasiyetin Sınırları)
Makine Epsilonu (ϵ): 1.0 sayısına eklendiğinde sonucu 1.0'dan farklı yapabilecek en küçük sayıdır. float için bu yaklaşık 1.19×10 üzeri -7'dir.
Yani float tipinde 1.0'a bundan daha küçük bir sayı eklerseniz, bilgisayar bunu algılayamaz ve sayıyı hala 1.0 zanneder.
ULP (Units at the Last Place): Birbirini takip eden iki ondalıklı sayı arasındaki fiziki "boşluktur"
Bilgisayardaki ondalıklı sayılar sürekli değildir; merdiven basamakları gibidir. Sayı büyüdükçe bu basamakların arasındaki boşluk (hata payı) da büyür.  

### Ondalıklı Sayıların Yıktığı Matematiksel Kurallar
En can alıcı noktalardan biri, lisede öğrendiğimiz matematik kurallarının ondalıklı sayılarda geçerli olmamasıdır:
- Birleşme Özelliği Yoktur (Not Associative): (a + b) + c işlemi C++'ta a + (b + c) işlemine eşit olmak zorunda değildir. Çünkü aralarda oluşan minik yuvarlama hataları sonuçları değiştirir. Örneğin 0.1 + (0.2 + 0.3) == (0.1 + 0.2) + 0.3 kodu C++'ta false döner..  
- Dağılma Özelliği Yoktur (Not Distributive): Aynı şekilde (a + b) * c işlemi her zaman (a * c) + (b * c) sonucunu vermez.  
- Tam sayıların (örneğin 16,777,217) kendisi, bazen float formatına çevrildiğinde değer kaybına uğrayıp 16,777,216'ya düşebilir. Büyük sayılarda int, float'tan çok daha kesindir.  

### Yıkıcı İptal (Catastrophic Cancellation)
Bilgisayar bilimlerinde en korkulan hatalardan biridir. Ondalıklı sayılarla işlem yaparken anlamlı hassasiyetin (virgülden sonraki kısmın) kalıcı olarak yok olmasıdır. İki durumda gerçekleşir:
- Çok büyük bir sayı ile çok küçük bir sayıyı toplamak/çıkarmak: Örneğin bir döngüde 10,000,000 değerine sürekli 1 eklemeye çalışırsanız, makine epsilonu sınırına takıldığınız için 1 değeri "yutulur" ve sayı asla büyümez. Döngünüz sonsuz döngüye girer.  
- Birbirine çok yakın iki sayıyı birbirinden çıkarmak: İki yakın sayıyı (örneğin a = 1.0000001 ve b = 1.0000000) çıkardığınızda, baştaki bütün anlamlı büyük basamaklar yok olur ve elinizde sadece bilgisayarın rastgele oluşturduğu, hatalı "çöp" minik bitler kalır.  

### Ondalıklı Sayıları Karşılaştırma (Comparison) Kuralları
C++'ta hiçbir zaman iki ondalıklı sayı == veya != operatörleriyle kıyaslanmamalıdır
0.11f + 0.11f < 0.22f işleminin sonucu mantıken false olması gerekirken, C++'ta true çıkabilir
- Yanlış Çözüm: Aradaki farkı sabit bir epsilon değeriyle kıyaslamak (örneğin abs(a - b) < 0.0001). Bu çalışmaz çünkü sayılar çok büyüdüğünde hata payı da büyür ve 0.0001 kontrolü yetersiz kalır
- Doğru Çözüm (Bağıl Hata - Relative Error): İki sayı arasındaki farkı (a - b), sayıların birbirine olan oranına (veya maksimumlarına max(a, b)) bölerek bağıl bir hata oranı hesaplamak ve bu oranın izin verilen sınırın (relative error) altında kalıp kalmadığını kontrol etmektir.  

### Özel Veri Yapıları: struct, Bitfield ve union
- struct (Yapı): Farklı tiplerdeki değişkenleri tek bir paket (ünite) altında toplamanızı sağlar
 Hatta fonksiyonların içine bile sadece o fonksiyona özel yerel (local) struct'lar yazabilirsiniz
- Bitfield (Bit Alanları): Bazen değişkenlerin bellekte çok yer kaplamasını istemezsiniz.  
Örneğin bir sayı sadece 0 ile 255 arasında olacaksa, buna koca bir int (32 bit) vermek israftır. Bit alanları sayesinde int b3 : 8; yazarak bu değişkene sadece 8 bit alan ayırabilirsiniz. Veri sıkıştırmak için harikadır.
- union (Birlik): struct'a benzer ama çok tehlikeli ve özel bir farkı vardır: İçindeki tüm değişkenler bellekte aynı (üst üste binen) yeri paylaşır. Yani bir union'ın boyutu, içindeki en büyük değişkenin boyutu kadardır. İçindeki bir değişkeni değiştirdiğinizde, diğer değişkenin de bellekteki bitlerini bozmuş/değiştirmiş olursunuz. Modern C++ (C++17), bu tehlikeli yapı yerine çok daha güvenli olan std::variant'ı kullanmayı tavsiye eder.
