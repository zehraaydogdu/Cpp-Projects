## Yorum Satırı Kullanımı
Kodu yorum satırına almak için iki yöntemimiz vardır:

### 1. Yöntem (Blok Yorum)
Yazının başına ve sonuna `/*` ile `*/` eklenir. 
Örn: `/* Bu bir denemedir */`

### 2. Yöntem (Tek Satır)
Her satır için `//` kullanılmalıdır. 
Örn: `// Bu satır için ayrı`

---

## Bir satır aşağıya inme ve tab komutları
`\n` veya `endl` kullanılır.

```python
cout << "Hello, World!" << endl;
cout << "Hello, \tWorld!\n" ;
```

## Veri Tipleri
- `int`: Tam sayı - 4 byte 
- `float`: Ondalıklı sayılar - 4 byte
- `double`: Yine ondalıklı sayı ama daha büyük - 8 byte
- `bool`: True-false ataması - 2 byte
- `char`: Tek karakter (harf) - 1 byte
- `string`: Karakter dizisi (cümle) - degişken

```python
int sayi = 10;
float sayi2= 3.14;
double sayi3= 3.14159265358979323846;
char karakter = 'A';
string metin = "Merhaba, Dünya!";

bool isUpdated = true;
cout << isUpdated;
//Burda göreceğimiz çıktı 1 olacaktır.

//char veri tipinde eğer iki ifade kullanılırsa bize çıktı olarak sadece biri verilir. 
//çünkü char veri tipi tek bir karakteri temsil eder.
```
