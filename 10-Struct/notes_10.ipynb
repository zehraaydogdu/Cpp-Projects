## Struct  
Struct (yapı), birbiriyle ilişkili farklı türdeki değişkenleri tek bir çatı altında toplamanı sağlayan kullanıcı tanımlı bir veri tipidir.
Şu ana kadar int, string, float gibi hazır tipleri kullandık. Ancak bazen bir nesneyi tanımlamak için bu tekil tipler yetmez. Mesela bir "Öğrenci"yi tanımlamak için hem isme (string), hem numaraya (int), hem de ortalamaya (double) ihtiyaç vardır. Bunları ayrı ayrı değişkenlerde tutmak yerine bir struct içinde paketleriz.

```python
struct Calisan{
    string ad;
    string soyad;
    int yas;
    float maas;
};

int main(){
    Calisan ilk;

    cout << "Calisanin adini giriniz: ";
    cin >> ilk.ad;
    cout << "Calisanin soyadini giriniz: ";
    cin >> ilk.soyad;
    cout << "Calisanin yasini giriniz: ";
    cin >> ilk.yas;
    cout << "Calisanin maasini giriniz: ";
    cin >> ilk.maas;
    return 0;
}
```

## C++ Struct İle Fonksiyon Kullanımı

```python
struct Calisan{
    string ad;
    string soyad;
    int yas;
    float maas;
};

void displayData(Calisan ilk){
    cout << "Calisanin adi: " << ilk.ad << endl;
    cout << "Calisanin soyadi: " << ilk.soyad << endl;
    cout << "Calisanin yasi: " << ilk.yas << endl;
    cout << "Calisanin maasi: " << ilk.maas << endl;
}

int main(){
    Calisan ilk;

    cout << "Calisanin adini giriniz: ";
    cin >> ilk.ad;
    cout << "Calisanin soyadini giriniz: ";
    cin >> ilk.soyad;
    cout << "Calisanin yasini giriniz: ";
    cin >> ilk.yas;
    cout << "Calisanin maasini giriniz: ";
    cin >> ilk.maas;

    displayData(ilk);
    return 0;
}
```
