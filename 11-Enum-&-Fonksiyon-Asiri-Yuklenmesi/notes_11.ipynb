## Enum  
enum (Enumeration - Numaralandırma), bir grup sabit değeri anlamlı isimlerle bir arada tutmanı sağlayan kullanıcı tanımlı bir veri tipidir.
Kodumuzda "Haftanın Günleri"ni tutmak istiyoruz. Pazartesi için 0, Salı için 1, Çarşamba için 2 gibi sayılar kullanabiliriz. Ama kodun içinde sürekli if (gun == 2) yazarsak, bir süre sonra "2 neydi ya?" diye düşünmeye başlarız. enum kullanarak bu sayılara isim veririz ve kod çok daha okunabilir olur. Yani bu isimleri numaralandırıyoruz.

```python
enum hafta{
    Pazartesi,
    Sali,
    Carsamba,
    Persembe,
    Cuma,
    Cumartesi,
    Pazar
};

int main(){
    hafta bugun;
    bugun = Persembe;
    cout << "Persembe haftanin " << bugun+1 << ". gunudur.";
    return 0;
}
```

## Fonksiyonların Aşırı Yüklenmesi (Function Overloading)  
Aynı isme sahip birden fazla fonksiyonun, farklı parametre listeleriyle (farklı tipte veya sayıda argümanla) tanımlanabilmesidir.
Normalde bir dilde her fonksiyonun ismi eşsiz olmalıdır, ancak C++ gibi diller "Zaten ne göndereceğimi biliyorum, isme değil parametrelere bakarak hangisini çalıştıracağımı anlarım" der.
Tam sayılar için ayrı, ondalıklı sayılar için ayrı, üç tane sayı için ayrı fonksiyon isimleri uydurmak (toplaInt, toplaDouble, toplaUcSayi) çok yorucudur. Bunun yerine hepsine topla dersin, gerisini derleyiciye bırakırsın.
