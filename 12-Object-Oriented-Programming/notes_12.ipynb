## Object Oriented Programming  
C dilinde yapamazken C++'ta objeler oluşturabiliriz. Bu kısım bununla ilgilidir.

```python
class Oda{
    public: //Böyle yazdığımızda tüm fonksiyonlara ve değişkenlere erişilebilir olur.
        double kisaduvar;
        double uzunduvar;
        double yukseklik;

        double hesapla_metrekare(){
            return kisaduvar * uzunduvar;
        }
        double hesapla_hacim(){
            return kisaduvar * uzunduvar * yukseklik;
        }
};

int main(){
    Oda benimodam; //Oda sınıfından nesne oluşturduk. Oda sınıfının tüm özelliklerine benimodam nesnesi üzerinden erişebiliriz.
    
    benimodam.kisaduvar = 3.5; //Deger atarken nokta operatoru kullandık. benimodam nesnesinin kisaduvar özelliğine 3.5 değerini atadık.
    benimodam.uzunduvar = 4.0;
    benimodam.yukseklik = 2.5;
    cout << "Odamın metrekare alanı: " << benimodam.hesapla_metrekare() << endl;
    cout << "Odamın hacmi: " << benimodam.hesapla_hacim() << endl;
    return 0;
}

//Eğer public yerine private yazarsak, o zaman benimodam nesnesi üzerinden kisaduvar, uzunduvar ve yukseklik özelliklerine erişemeyiz.
//Bu durumda sadece hesapla_metrekare() ve hesapla_hacim() fonksiyonlarına erişebiliriz.
```

## Constructor Yapısı  
Nesne doğduğu an onun özelliklerini belirlemeye yarar. Yani diyelim oyun yazıyoruz yeni karakter oluşturduk bu classtan ve   
constructor yapısıyla canına 100 atamıştık. Her yeni karakter oluştuğunda 100 canla başlamış olur olay bu.

```python
class Kisi{
    private:
    int yas;
    public:
    Kisi():yas(40){
        cout << "Zehranin yasi: " << yas;
    }
};

int main(){
    Kisi zehra; //Kisi sınıfının constructor'ı çalışacak ve yas değişkenine 40 değeri atanacak.
    return 0;
}


//Ek bilgi: Parameteli constructorlar parametresiz constructorları ezer.
//Eğer Kisi sınıfında parametresiz constructor ve parametreli constructor varsa, parametreli constructor çağrıldığında 
//parametresiz constructor çalışmaz. Parametreli constructor, parametresiz constructor'ı ezer ve sadece kendisi çalışır. 
```
