## Inheritance  
Ana class ve çocuk classlar olur. Bu çocuk classlar ana classın özelliklerini taşır ve de kendilerine özel özellikleri de bulunur.

```python
class Hayvan{
    public:
    void yemekYer(){
        cout << "Hayvan yemek yer." << endl;
    }
    void uyur(){
        cout << "Hayvan uyur." << endl;
    }
};

class Kopek : public Hayvan{
    public:
    void havlar(){
        cout << "Kopek havlar." << endl;
    }
    void dostOlur(){
        cout << "Kopek dost olur." << endl;
    }
};

int main(){
    Kopek k1;
    k1.yemekYer();
    k1.uyur();
    k1.havlar();
    k1.dostOlur(); // Kopek sınıfına özgü fonksiyonları çağırabiliriz.

}

//public yerine protected veya private kullanırsak, Kopek sınıfı Hayvan sınıfının fonksiyonlarına erişemez. 
//Bu durumda Kopek sınıfı sadece kendi fonksiyonlarını kullanabilir, ancak Hayvan sınıfının fonksiyonlarına erişemez.
```
