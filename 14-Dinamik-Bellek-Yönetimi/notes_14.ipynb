## Dinamik Bellek Yönetimi  
New ile bellekten bir yer ayrılırken delete ile silinir ayrılan yer.  
Kullanımı şu şekildedir:

```python
int main(){
    int *sayi;
    sayi = new int;
    *sayi = 10;

    cout << *sayi << endl;
    return 0;
}
```

Başka bir şekilde de yazabiliriz:

```python
int main(){
    int *sayi = new int{5};
    cout << *sayi << endl;
    return 0;
}
```

delete komutunu ise şöyle kullanırız:

```python
int main(){
    int *sayi = new int{5};
    cout << *sayi << endl;
    delete sayi; //Eğer istersek bu komuttan sonra sayıya sayi = nullptr; diyebiliriz. 
    //Böylece sayi değişkeni artık hiçbir adresi göstermeyecektir.
    return 0;
}
```
