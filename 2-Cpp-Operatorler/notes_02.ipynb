## Operatörler 
C dilinden farkı yoktur:  
- Toplama = +  
- Çıkarma = -  
- Çarpma = *  
- Bölme = /  
- Mod Operatörü = %  
- Arttırma/Azaltma = ++ veya --  

**Aritmetik Operatörler**  
+=, -=, *=, /=  

**Karşılaştırma Operatörleri**  
!=, >, <, >=, <=, ==  

**Mantıksal Operatörler**  
&& = Ve  
|| = Ya Da  
! = Değil  

**Ternary Operatörü**  
Koşula göre atama yapar.
Kullanımı şu şekildedir:  
`int sayi;`  
`sayi=10>15 ? 8:7`  
*Demek istiyor ki 10 sayısı 15ten büyükse ekrana 8i eğer değilse ekrana 7yi bastır.*  
*Bizim çıktımız 10, 15ten küçük olduğu için 7dir.*  

**i++ ve ++i arasındaki fark:**  

`int sayi1=10, sayi2=10;`  
`sayi2 = ++sayi1;`  
*denilip ekrana yazdırılırsa alacağımız çıktıda önce sayi1 arttırılır daha sonra atama yapılır.*  
*yani sonuç olarak ekranda sayi1=11, sayi2=11 olduğu görülür.*   

`int sayi1=10, sayi2=10;`  
`sayi2 = sayi1++;`  
*denilip ekrana yazdırıldığında ise önce atama yapılır daha sonra sayi1 arttırılır.*
*yani bu defa çıktımızda sayi1=11, sayi2=10 olarak gözükür.*

## Casting İşlemi (Tip Dönüşümü)   


```c++
#include <iostream>
using namespace std;
int main(){
    int sayi1=10, sayi2=4;
    cout<<"Sayi1 in Sayi2 ye bolumu: "<<(float)sayi1/sayi2;
    return 0;
} 
```

## Sizeof Kullanımı  
Sizeof girilen ifadelerin bellekte ne kadar yer kapladığını bize gösterir.

```c++
int main(){
    cout<<"char "<<sizeof(char)<<" byte yer kaplar.";
}
//Çıktısı: char 1 byte yer kaplar.
```
