# Döngüler

## For Döngüsü  
C dilindekiyle aynı yapıdadır.  

`for(ilk deger atamasi; koşul; arttırma veya azaltma durumu){`  
`}`  

## While Döngüsü  
İçindeki koşul doğru olduğu sürece çalışır. Yine c diliyle aynı.  

## Do-While Döngüsü  
Whileda koşulun içi doğru değilse hiç başlanmaz program bir şeyler oluşturmaya.  
Ama do-while kullanırsak program bir defa çalışır başlangıçta. Farkları budur.

---

## Break-Continue  
- Break: Programın çalışmasını sonlandırır.  
- Continue: Koşul sağlanınca o bloğu atlar.  

```c++
//Continue Kullanımı  
for(i=1; i<=10; i++){  
    if(i==2){  
       continue;
    }  
    cout<<i<<endl;
}  

//Çalıştırıldığında 2 değerini atlayarak çıktı verir. 1-3-4-5-6-7-8-9-10
```
