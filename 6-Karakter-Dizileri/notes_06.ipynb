### Getline
!! String kütüphanelerini entegre etmeyi unutmamak gerekir.  

```python
int main(){
    string message;
    cout << "Lutfen mesajınızı giriniz: ";
    getline(cin, message);

    cout << "Girdiğiniz mesaj: " << message << endl;
    return 0;
}
```

cin ve getline arasındaki fark da şu mesela yazdığın şeyde boşluk olunca cin gerisini almaz. Ama getline alır.  
Yani boşluklu ifade yazınca getline ile almamız lazım. Bir de cin ile tüm veri tiplerini alabilirken getline sadece string.

### Karakter Dizisinin Boyunu Bulmak

```python
int main(){
    string message = "Hello";
    int length = message.size();

    cout << length << endl;
    return 0;
}
```

```python
//İki stringi karşılaştırmak için compare() fonksiyonunu kullanabiliriz.   
//Bu fonksiyon, iki stringi sözlük sırasına göre karşılaştırır ve aşağıdaki sonuçları döndürür.

int main(){
    string ilk = "osman";
    string ikinci = "gazi";
    int sonuc= ilk.compare(ikinci);
    if sonuc == 0 {
        cout << "iki string birbirine eşittir" << endl;
    } else if (sonuc < 0) {
        cout << "ilk string ikinci stringden küçüktür" << endl;
    } else {
        cout << "ilk string ikinci stringden büyüktür" << endl;
    }

    return 0;
}
```

```python
//Stringi tam sayıya dönüştürmek için stoi() fonksiyonunu kullanabiliriz.
//int bir ifadeyi stringe dönüştürmek için to_string() fonksiyonunu kullanabiliriz.

int main(){
    string MyNumber = "12345";
    int number = stoi(MyNumber);
    cout << number;

    return 0;
}
```

```python
//strlen() fonksiyonu, bir karakter dizisinin uzunluğunu hesaplamak için kullanılır.

int main(){
    char message[20]= "Hello World";
    int number = strlen(message);
    cout << number << endl;
    return 0;"
}
```

```python
//at() fonksiyonu, bir stringin belirli bir konumundaki karakteri döndürür. 
//Örneğin, message.at(0) ifadesi, message stringinin ilk karakterini döndürür.

int main(){
    string message = "Hello";
    char firstChar = message.at(0);
    cout << firstChar << endl;
    return 0;
}
```

```python
//find() fonksiyonu, bir string içinde belirli bir karakterin veya alt stringin konumunu bulmak için kullanılır.

int main(){
    string message = "Hello World";
    int index = message.find("World");
    cout << index << endl;
    return 0;
}
```

```python
//insert() fonksiyonu, bir stringin belirli bir konumuna başka bir string eklemek için kullanılır.

int main(){
    string message = "Hello World";
    message.insert(5, " Beautiful");
    cout << message << endl;
    return 0;
}

//Çıktı: Hello Beautiful World
```
