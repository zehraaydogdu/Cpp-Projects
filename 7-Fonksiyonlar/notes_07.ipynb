## Fonksiyonlar  

Geriye değer döndürmeyen fonksiyonlar "void"lerdir. Kullanımı şu şekildedir:

```python
void kareal(){
    int sayi, karesi;
    cout << "Lutfen bir sayi giriniz: ";
    cin >> sayi;
    karesi = sayi * sayi;
    cout << "Girdiginiz sayinin karesi: " << karesi << endl;
}

int main(){
    kareal();
    return 0;
}

//kareal(sayi) olarak da kullanılabilir.  
//Bu durumda, fonksiyonun parametre alması gerekir.  
void kareal(int sayi){
    int karesi = sayi * sayi;
    cout << "Girdiginiz sayinin karesi: " << karesi << endl;
}
```

Geriye değer döndüren fonksiyonlar döndürdüğümüz değer tipindedir. Örnek vermemiz gerekirse:

```python
int kareal(int a){
    int karesi;
    karesi = a * a;
    return karesi;
}

int main(){
    int sayi;
    cout << "Lutfen bir sayi giriniz: ";
    cin >> sayi;
    int sonuc = kareal(sayi);
    cout << "Girdiginiz sayinin karesi: " << sonuc << endl;
    return 0;
}  

//Burdaki int a ifadesi önemli değildir, herhangi bir isim kullanılabilir. Önemli olan, fonksiyonun bir tam sayı parametresi almasıdır.
```

### Fonksiyonları Referans İle Çağırmak

```python
void bastir(int *sayi){
    *sayi = 9;
    cout << "Fonksiyonun icindeki sayi: " << *sayi << endl;
}

int main(){
    int x = 5;
    bastir(&x);
    cout << "Main içerisindeki sayi: " << x << endl;
    return 0;
}

//Burda iki çıktı da 9 olacaktır, çünkü fonksiyon içinde sayi değişkeninin değeri değiştirilmiştir. 
//Ve bu değişiklik main fonksiyonundaki x değişkenine de yansımıştır.
//Main içerisindeki xi bastırırken cout kısmında * kullanılmaz, çünkü x zaten bir tam sayı değişkenidir.
//Ve doğrudan değeri bastırılabilir.
```

### Random Sayı Üretme

```python
#include <time.h>
int main(){
    srand(time(0)); //Rastgele sayı üretmek için srand() fonksiyonunu kullanırız. 
    //time(0) ifadesi, mevcut zamanı saniye cinsinden döndürür ve bu değeri rastgele sayı üretmek için kullanırız.
    int randomNumber = rand() % 100 + 1; //rand() fonksiyonu, 0 ile RAND_MAX arasında rastgele bir tam sayı üretir. 
    //% 100 ifadesi, üretilen sayıyı 0 ile 99 arasında sınırlamak için kullanılır. + 1 ifadesi ise sonucu 1 ile 100 arasında yapar.
    cout << "Rastgele sayi: " << randomNumber << endl;
    return 0;
}
```
