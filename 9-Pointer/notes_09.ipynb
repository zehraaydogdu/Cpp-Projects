## Pointer
Adres ve adresin gösterdiği değerdir. Bir değerin kendisini değil, o değerin bellekteki (RAM) adresini tutan bir değişkendir.  

Pointer dünyasında anlaman gereken iki ana operatör vardır:
- (Adres Operatörü) & : Bir değişkenin bellekteki adresini verir. "Sen nerede oturuyorsun?" sorusudur.  
- (Değer/Yönlendirme Operatörü) * : Bir adreste hangi değerin olduğunu verir. "Bu adresteki evde kim yaşıyor?" sorusudur.  

Hem pointer hem de normal veri tipi yan yana tanımlanabilir:
``` int a, *b;```

````python
int main(){
    int a = 5;
    int *y;
    y = &a;

    cout << "A degerinin adresi: "<< &a << endl;
    cout << "Y degeri: " << y << endl;
    cout << "Y degerinin gosterdigi deger: " << *y << endl;
    cout << "A degeri: " << a << endl;

    return 0;
}

//Burada ilk çıktıda a değişkeninin adresi yazılır. Buna 1005 diyelim.  
//İkinci çıktıda y değişkeninin değeri yazılır. Bu da a değişkeninin adresi olduğu için 1005 olur.
//Üçüncü çıktıda y değişkeninin gösterdiği değeri yazarız. Bu da a değişkeninin değeri olduğu için 5 olur.
//Dördüncü çıktıda a değişkeninin değeri yazılır, bu da 5 olur.
````

````python
int main(){
    int *sayim;
    int sayi[5];

    sayim = sayi; //sayim = &sayi[0]; ile aynı anlama gelir. Yani sayim dizinin ilk elemanının adresini tutar.
}
````

### Fonksiyonları Referans İle Çağırma

````python
void degeryolla(int *x){
    *x = 30;
    cout << "X degerim: " << *x << endl;
}

int main(){
    int x = 20;
    degeryolla(&x);
    cout << "X degerim: " << x << endl;
}

//Bu örnekte, degeryolla fonksiyonu bir pointer parametresi alır. main fonksiyonunda x değişkeni tanımlanır ve değeri 20 olarak atanır. 
//Daha sonra degeryolla fonksiyonuna x değişkeninin adresi (&x) gönderilir. 
//Ve degeryolla fonksiyonu x değişkeninin değerini 30 olarak değiştirir.
//Çıktıda önce degeryolla fonksiyonunun içindeki x değişkeninin değeri yazılır, bu da 30 olur.
//Daha sonra main fonksiyonunda x değişkeninin değeri yazılır, bu da 30 olur.
````
